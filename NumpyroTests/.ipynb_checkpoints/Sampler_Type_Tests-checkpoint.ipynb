```python
import jax
import jax.numpy as jnp
import numpyro
import numpy as np
from chainconsumer import ChainConsumer
import matplotlib.pylab as plt
```

```python
N = 10
w_true = 1
X = np.linspace(0,np.pi*(N-1), N )
Y = np.cos(w_true*X)
E = np.random.poisson(lam=1, size=N) * 0.1
Y += np.random.randn() * E
```

```python
Xplot = np.linspace(0,np.pi*(N-1), N*100 )
Yplot = np.cos(w_true*Xplot)
plt.errorbar(X, Y, E, fmt='none')
plt.plot(Xplot,Yplot,'--',lw=0.5)
plt.show()
```

```python
def model(X,Y,E):
    A = numpyro.sample('A',numpyro.distributions.Uniform(0, 10))
    w = numpyro.sample('w',numpyro.distributions.Uniform(0, 10))
    phi = numpyro.sample('phi',numpyro.distributions.Uniform(0, np.pi))
    
    with numpyro.plate('data',len(X)):
        ypred = A*jnp.cos(w*X+phi)
        numpyro.sample('y', numpyro.distributions.Normal(ypred, E), obs=Y)
        
sampler = numpyro.infer.MCMC(
        numpyro.infer.NUTS(model),
        num_warmup=100,
        num_samples=100,
        num_chains=1,
        progress_bar=True,
    )

sampler.run(jax.random.PRNGKey(0), X, Y, E)
```
